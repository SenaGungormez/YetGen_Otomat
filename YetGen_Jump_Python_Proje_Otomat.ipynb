```python
import time

class Otomat():
  def __init__(self, otomatDurum = "Kapalı", miktar = "Küçük", menu = ["Espresso", "Sıcak Çikolata", "Latte", "Mocha"], 
    fiyatListe = {
      "Espresso" : [15,20,25],
      "Sıcak Çikolata" : [20,25,30],
      "Latte" : [25,30,35],
      "Mocha" : [25,30,35]
   }, 
   ekstraListe  = {
       "Şeker" : 1,
       "Espresso Shot" : 10
   },
   surupListe = {
       "Vanilya" : 2,
       "Karamel" : 2,
       "Fındık" : 2,
       "Çilek" : 2
    } ):
    self.otomatDurum = otomatDurum
    self.menu = menu
    self.miktar = miktar
    self.fiyatListe = fiyatListe
    self.ekstraListe = ekstraListe
    self.surupListe = surupListe 
  
  def otomatAc(self):
    if (self.otomatDurum == "Kapalı"):
      print("Otomat açılıyor.")
      time.sleep(2)
      self.otomatDurum = "Açık"
      print("Otomat açıldı.")
    else:
      print("Otomat zaten açık.")

  def otomatKapat(self):
    if (self.otomatDurum == "Açık"):
      print("Otomat kapatılıyor.")
      time.sleep(2)
      self.otomatDurum = "Kapalı"
      print("Otomat kapatıldı.") 
    else:
      print("Otomat zaten kapalı.")  

  def kahveEkle(self):
    if (self.otomatDurum == "Açık"):
     yeniFiyatListe = []
     userInput = input("Eklemek istediğiniz kahvenin adını giriniz: ")
     userInput = userInput.title()
     self.menu.append(userInput)
     yeniKucukFiyat = int(input("Yeni eklediğiniz kahvenin küçük boy fiyatını giriniz: "))
     yeniFiyatListe.append(yeniKucukFiyat)
     yeniOrtaFiyat = int(input("Yeni eklediğiniz kahvenin orta boy fiyatını giriniz: "))
     yeniFiyatListe.append(yeniOrtaFiyat)
     yeniBuyukFiyat = int(input("Yeni eklediğiniz kahvenin büyük boy fiyatını giriniz: "))
     yeniFiyatListe.append(yeniBuyukFiyat)       
     self.fiyatListe[userInput] = yeniFiyatListe
     print(self.fiyatListe)
    else:
      print("Otomatı Açtıktan Sonra Deneyiniz.")

  def displayMenu(self):
    print("""
    --------- OTOMAT ---------
    1. Otomat Aç
    2. Otomat Kapat
    3. Kahve Ekle
    4. Kahve Satın Al
    """)

  def displayCoffeeMenu(self):
    for item in list(self.fiyatListe.keys()):
      print(f"{item.upper()} : {self.fiyatListe[item]} TL")

  def kahveSecme(self):
    kahve = input("Almak istediğiniz kahvenin adını giriniz: ")
    kahve = kahve.title()
    boyut = int(input("\n1. Küçük\n2. Orta\n3. Büyük\nKahvenizin boyutunun numarasını giriniz: "))
    fiyat = 0
    if boyut == 1:
      fiyat = self.fiyatListe[kahve][0]
    elif boyut == 2:
      fiyat = self.fiyatListe[kahve][1]
    elif boyut == 3:
      fiyat = self.fiyatListe[kahve][2]
    print()

    for item in list(self.ekstraListe.keys()):
      print(f"{item.upper()} : {self.ekstraListe[item]} TL")
    print()
    ekstra = input("İstedğiniz ekstra ürünü giriniz (Ekstra ürün istemiyorsanız 0'a basınız.): ")
    ekstra = ekstra.title()
    for item in list(self.ekstraListe.keys()):
      if item == ekstra:
        fiyat += self.ekstraListe[item]
      elif ekstra == 0:
        print(f"Ekstra ürün konulmuyor.")
    print()

    for item in list(self.surupListe.keys()):
      print(f"{item.upper()} : {self.surupListe[item]} TL")
    print()
    surup = input("İstediğiniz şurubu giriniz: (Şurup istemiyorsanız 0'a basınız.): ")
    surup = surup.title()          
    for item in list(self.surupListe.keys()):
      if item == surup:
        fiyat += self.surupListe[item]
      elif surup == 0:
        print(f"Şurup konulmuyor.")  
    print()
    check = input(f"Ödemeniz gereken toplam fiyat: {fiyat} TL \nSatın almayı onaylıyor musunuz? (E/H): ")
    check = check.upper()
    if check == "E":
      print("Satın alma işlemi onaylandı.\nTEŞEKKÜRLER.")
    elif check == "H":
      print("Satın alma işlemi iptal edildi.")  




otomat = Otomat()

otomat.displayMenu()
while(True):


  userInput = input("\nYapmak istediğiniz işlemin numarasını giriniz: ")
  print()

  if (userInput == "1"):
    otomat.otomatAc()
  elif (userInput == "2"):
    otomat.otomatKapat()  
  elif (userInput == "3"):
    otomat.kahveEkle()  
  elif (userInput == "4"):
    otomat.displayCoffeeMenu()
    otomat.kahveSecme()
```
